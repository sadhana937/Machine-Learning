## Polynomial Regression

Preparing data which does not have a linear relationship

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pageSpeeds = np.random.normal(3.0,1.0,100)
purchaseAmounts = np.random.normal(50.0,10.0,100) / pageSpeeds
plt.scatter(pageSpeeds, purchaseAmounts)
plt.show()
```

Using the polyfit function from numpy we can construct nth-degree polynomial model:
    

```python
x = np.array(pageSpeeds)
y = np.array(purchaseAmounts)

p = np.poly1d(np.polyfit(x,y,4)) # creating a polynomial object where n=4
```

```python
xp = np.linspace(0,7,100)
plt.scatter(x, y)
plt.plot(xp, p(xp), c='r')
plt.show()
```

Finding the r-squared error

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p(x))
print(r2)
```
