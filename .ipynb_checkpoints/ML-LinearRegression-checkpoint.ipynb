## Linear Regression

Fabricating some data that shows a linear relationship between page speed and amount purchased.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
pageSpeeds = np.random.normal(3.0,1.0,10000)
purchaseAmounts = 100 - (pageSpeeds + np.random.normal(0,0.1,10000)) * 3

plt.scatter(pageSpeeds, purchaseAmounts)
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmounts)
R_squared_value = r_value ** 2
print(R_squared_value)
```

R-squared value represents a really good fit.Plotting the predicted vs. observed values using the slope and intercept.

```python
def predict(x):
    return slope * x + intercept

predicted_line = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmounts)
plt.plot(pageSpeeds, predicted_line, c='r')
plt.show()
```
